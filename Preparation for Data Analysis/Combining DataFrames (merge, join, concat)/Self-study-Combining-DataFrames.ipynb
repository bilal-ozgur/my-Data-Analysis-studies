# Combining DataFrames

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

## Some examples DataFrames

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']}
                    )
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']}
                    ) 
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']}
                   )
```

```python
df4 = pd.DataFrame({'C': ['C12', 'C13', 'C14', 'C15'],
                    'A': ['A12', 'A13', 'A14', 'A15'],
                    'D': ['D12', 'D13', 'D14', 'D15'],
                    'B': ['B12', 'B13', 'B14', 'B15']}
                    )
```

```python
display(df1,df2,df3,df4)
```

# Append

```python
df1.append(df3)
```

```python
df1.append(df3,ignore_index=True)
```

```python
df1.append(df4)
```

# Concatenation

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],ignore_index=True)
```

```python
pd.concat([df1.iloc[:,:2],df2.iloc[:,1:],df3])
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
pd.concat([df1.iloc[:2],df2.iloc[1:],df3],axis=1)
```

```python
df4 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])

df5 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7]) 

df6 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
```

```python
display(df4,df5,df6)
```

```python
pd.concat([df4,df5,df6],axis=1)
```

```python
pd.concat([df4,df5,df6],axis=1,ignore_index=True)
```

```python
pd.concat([df4,df5,df6],axis=1,join='inner')
```

```python
pd.concat([df4,df5.reset_index(),df6.reset_index()],axis=1)
```

```python
pd.concat([df4,df5.reset_index(drop=True),df6.reset_index(drop=True)],axis=1)
```

```python
pd.concat([df4,df5.reset_index(drop=True),df6.reset_index(drop=True)],axis=1,join='inner')
```

```python
pd.concat([df4,df5,df6])
```

# Merging

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})   
```

```python
display(left, right)
```

```python
pd.merge(left,right,how='inner',on='key')
```

```python
left2 = pd.DataFrame({'key': ['K0', 'K1', 'K4', 'K5'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']}) 
```

```python
display(left2, right2)
```

```python
pd.merge(left2,right2,how='inner',on='key')
```

```python
pd.merge(left2,right2,how='outer',on='key')
```

```python
pd.merge(left2,right2,how='left',on='key')
```

```python
pd.merge(left2,right2,how='right',on='key')
```

```python
left3 = pd.DataFrame({'key': ['K0', 'K0', 'K1', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right3 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K2'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
display(left3, right3)
```

```python
pd.merge(left3,right3,how='inner',on='key')
```
