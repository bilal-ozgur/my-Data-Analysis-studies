## FUNCTIONAL METHOD & OBJECT-ORIENTED METHOD

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]

salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]

salary_2 = [45372, 48876, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
```

```python
# FUNCTIONAL METHOD

plt.plot(age,salary)
plt.title('Salary by Age')
plt.xlabel('Age')
plt.ylabel('Salary');
```

```python
plt.plot(age,salary_2)
plt.title('Salaries by Age')
plt.xlabel('Age')
plt.ylabel('Salary_2');
```

```python
## OBJECT ORIENTED METHOD

fig, ax = plt.subplots()

ax.plot(age,salary, 'r')
ax.set_title('Salaries by Age')
ax.set_xlabel('Age')
ax.set_ylabel('Salary');
```

```python
fig ,ax = plt.subplots()

ax.plot(age,salary_2,'r')
ax.plot(age,salary,'b')
ax.set_title('Salaries by Age')
ax.set_xlabel('Age')
ax.set_ylabel('Salary_02');
```

## Multiple-Line in the same graph

```python
# F
plt.plot(age,salary)
plt.plot(age,salary_2)
plt.title('Salaries by Age')
plt.xlabel('Age')
plt.ylabel('Salary_02');
```

```python
# OOM
fig, ax = plt.subplots()
ax.plot(age,salary,'b')
ax.plot(age,salary_2,'r')
ax.set_title('Salaries by Ages')
ax.set_xlabel('Age')
ax.set_ylabel('Salary');
```

# Subplots

```python
plt.plot(age,salary,'r')
plt.plot(age,salary_2,'b')

plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary by Age');
```

```python
# FUNCTIONAL METHOD

plt.subplot(2,1,1)
plt.plot(age,salary)

plt.subplot(2,1,2)
plt.plot(age,salary_2);
```

```python
plt.subplot(1,2,1)
plt.plot(age,salary)

plt.subplot(1,2,2)
plt.plot(age,salary_2)

plt.tight_layout();
```

```python
#OBJECT ORIENTED METHOD
```

```python
# vertical
fig, ax = plt.subplots(nrows=1,ncols=2)

ax[0].plot(age,salary,'r')
ax[0].set_title('First plot')

ax[1].plot(age,salary_2,'b')
ax[1].set_title('Second plot')

plt.tight_layout();
```

```python
# horizontal
fig, ax = plt.subplots(nrows=2,ncols=1)

ax[0].plot(age,salary,'r')
ax[0].set_title('First plot')

ax[1].plot(age,salary_2,'b')
ax[1].set_title('Second plot')

plt.tight_layout();
```

```python
#FUNCTIONAL

fig = plt.figure()

plt.subplot(2,2,1)
plt.plot(age,salary,'r')

plt.subplot(2,2,2)
plt.plot(age,salary_2,'g')

plt.subplot(2,2,3)
plt.plot(age,salary,'b')

plt.subplot(2,2,4)
plt.plot(age,salary,'y')

# automatic:
# plt.tight_layout();

#manuel:
fig.subplots_adjust(left=None,
                   bottom=None,
                   right=None,
                   top=None,
                   wspace=0.9,
                   hspace=0.5);
```

```python
# OOM

fig, ax = plt.subplots(2, 2, figsize=(8, 5), sharey=True, sharex=True)

ax[0, 0].plot(age, salary, "r")
ax[0, 0].set_title("salary by age")
ax[0, 1].plot(age, salary_2)
ax[0, 1].set_title("salary_2 by age")
ax[1, 0].plot(age, salary, "g")
ax[1, 0].set_title("salary by age")
ax[1, 1].plot(age, salary_2, "y")
ax[1, 1].set_title("salary_2 by age")

fig.subplots_adjust(left=0.1,
                    bottom=0.2,
                    right=None,
                    top=None,
                    wspace=0.9,
                    hspace=0.5)
#plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(2,3)

ax[1][2].plot(age,salary)

plt.tight_layout();
```

```python
fig , ax = plt.subplots(2,3)

ax[1][2].plot(age,salary)
ax[0][1].plot(age,salary_2)

plt.tight_layout();
```

## How to manipulate figure size

```python
#FUNCTIONAL

plt.figure(figsize=(16,8))

plt.plot(age,salary)

plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary by Age');
```

```python
plt.figure(figsize=(8,4))

plt.plot(age,salary)

plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary by Age');
```

```python
# OOM METHOD

fig, ax = plt.subplots(figsize=(16,8))

ax.plot(age,salary,'r')

ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_title('Salaries by Age');
```
