# Data Visualization with Seaborn

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip']);
```

```python
sns.scatterplot(data = tips, x="total_bill", y="tip");
```

```python
tips.head()
```

### hue

```python
sns.scatterplot(data = tips, x="total_bill", y="tip",hue='sex');
```

### s

```python
sns.scatterplot(data = tips, x="total_bill", y="tip",hue='sex',s=200);
```

### size & sizes

```python
sns.scatterplot(data = tips, x="total_bill", y="tip",size='size',sizes=[10,30,50,70,90,110]);
```

### style

```python
tips.head()
```

```python
sns.scatterplot(data = tips, x="total_bill", y="tip",style='smoker');
```

### alpha

```python
sns.scatterplot(data = tips, x="total_bill", y="tip",s=200, alpha=0.4);
```

### Exporting figures

```python
sns.scatterplot(data = tips, x="total_bill", y="tip",s=200, alpha=0.4);

plt.savefig('sample.jpg')
```

# Rugplot

```python
sns.rugplot( x= tips['total_bill'] );
```

```python
sns.rugplot( x= tips['total_bill'],height=0.5)
```
